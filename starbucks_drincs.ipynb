```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
```

```python
df = pd.read_csv("D:\codingenv\deploma\project\starbucks_drinks.csv")
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.replace(" ","_").str.strip("_")
df.columns
```

```python
df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)',
       'Calcium_(%_DV)', 'Iron_(%_DV)']] = df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)',
       'Calcium_(%_DV)', 'Iron_(%_DV)']].replace('%', '', regex=True)
df.head()
```

```python
df[[ 'Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)',
       'Calcium_(%_DV)', 'Iron_(%_DV)']] = df[[ 'Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)',
       'Calcium_(%_DV)', 'Iron_(%_DV)']].astype(float)
df.head()
```

```python
df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)',
       'Calcium_(%_DV)', 'Iron_(%_DV)']] = df[[ 'Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)',
       'Calcium_(%_DV)', 'Iron_(%_DV)']]/100
df.head()
```

```python
df['Total_Fat_(g)'] = pd.to_numeric(df['Total_Fat_(g)'], errors='coerce')
df.head()
```

```python
df['Caffeine_(mg)'] = pd.to_numeric(df['Caffeine_(mg)'], errors='coerce')
df.head()
```

```python
df['Total_Fat_(g)'] = df['Total_Fat_(g)'].fillna(0).astype('float')
df['Total_Fat_(g)'] = df['Total_Fat_(g)']/100
df['Caffeine_(mg)'] = df['Caffeine_(mg)'].fillna(0).astype('int64')
df.head()
```

### What is the impact of different beverage preparations on the vitamin A , C , Calcium and Iron content?

```python
df.groupby('Beverage_prep')[['Vitamin_A_(%_DV)','Vitamin_C_(%_DV)','Calcium_(%_DV)',
                             'Iron_(%_DV)']].mean().reset_index().sort_values(by='Beverage_prep')
```

```python
vitamin_mineral_means = df.groupby('Beverage_prep')[['Vitamin_A_(%_DV)','Vitamin_C_(%_DV)','Calcium_(%_DV)',
                                                     'Iron_(%_DV)']].mean().reset_index().sort_values(by='Beverage_prep')
vitamin_mineral_melted = vitamin_mineral_means.melt(id_vars='Beverage_prep', var_name='Vitamin/Mineral', value_name='Mean (%DV)')
plt.figure(figsize=(10, 6))
sns.barplot(x='Beverage_prep', y='Mean (%DV)', hue='Vitamin/Mineral', data=vitamin_mineral_melted)
plt.xlabel('Beverage Preparation')
plt.ylabel('Mean (%DV)')
plt.title('Mean Vitamin and Mineral Content by Beverage Preparation')
plt.legend(title='Vitamin/Mineral')
plt.show()
```
